```python
import torch
import torch.nn as nn
from torchvision import models, transforms
from PIL import Image
```

```python

# Define the feature extraction model
class ImageFeatureExtractor(nn.Module):
    def __init__(self, output_dim):
        super(ImageFeatureExtractor, self).__init__()
        resnet = models.resnet50(pretrained=True)
        # Remove the FC layer
        self.cnn_backbone = nn.Sequential(*list(resnet.children())[:-1])
        # Optional: Map to desired output dimension
        self.visual_fc = nn.Linear(resnet.fc.in_features, output_dim)

    def forward(self, images):
        # Extract features using ResNet (without FC)
        visual_feats = self.cnn_backbone(images).squeeze(-1).squeeze(-1)
        # Map features to desired dimension
        visual_feats = self.visual_fc(visual_feats)
        return visual_feats

# Preprocess the image
def preprocess_image(img_path):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),  # ResNet expects 224x224 input
        transforms.ToTensor(),         # Convert image to tensor
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],  # Normalize as per ImageNet
            std=[0.229, 0.224, 0.225]
        )
    ])
    image = Image.open(img_path).convert("RGB")
    image = transform(image)
    return image.unsqueeze(0)  # Add batch dimension

# Example Usage
if __name__ == "__main__":
    # Path to an input image
    img_path = "/content/toxic5.jpg"  # Replace with your image path

    # Initialize the feature extractor
    output_dim = 768  # Example output dimension
    feature_extractor = ImageFeatureExtractor(output_dim)

    # Preprocess the image
    image_tensor = preprocess_image(img_path)

    # Extract features
    feature_extractor.eval()  # Set the model to evaluation mode
    with torch.no_grad():
        visual_features = feature_extractor(image_tensor)

    print("Extracted Visual Features Tensor Shape:", visual_features.shape)
    print("Extracted Visual Features Tensor:", visual_features)
```

```python


# Define a simple decoder
class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(512, 256, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1),
            nn.Tanh()  # Normalize to [-1, 1] for image
        )

    def forward(self, x):
        return self.decoder(x)

# Assume the tensor is the output of ResNet (e.g., 512 channels, 1x1 spatial resolution)
# Expand dimensions if necessary
tensor = torch.randn(1, 512, 1, 1)  # Example tensor

# Instantiate the decoder
decoder = Decoder()

# Generate the reconstructed image
reconstructed_img = decoder(tensor)

# Visualize the reconstructed image
reconstructed_img = reconstructed_img.detach().squeeze().permute(1, 2, 0).cpu().numpy()
reconstructed_img = (reconstructed_img + 1) / 2  # Rescale to [0, 1]

plt.imshow(reconstructed_img)
plt.axis('off')
plt.show()
```

```python
import torch
import torchvision.transforms as transforms
from torchvision import models
from PIL import Image
import matplotlib.pyplot as plt

# Load an image
image_path = "/content/toxic9.jpg"  # Replace with your image path
image = Image.open(image_path).convert("RGB")

# Define the image transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to ResNet input size
    transforms.ToTensor(),  # Convert to tensor
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1]
])

# Apply transformation
tensor = transform(image)
tensor = tensor.unsqueeze(0)  # Add batch dimension

# Load pretrained ResNet model (ResNet-18 for simplicity)
resnet = models.resnet18(pretrained=True)

# Remove the fully connected (FC) layer
resnet = torch.nn.Sequential(*list(resnet.children())[:-1])  # Removes the final FC layer

# Set the model to evaluation mode
resnet.eval()

# Pass the image tensor through ResNet to get features
with torch.no_grad():
    features = resnet(tensor)

# The output `features` is of shape (1, 512, 1, 1), which is a feature map of the last layer
# Let's squeeze it to remove singleton dimensions
features = features.squeeze().cpu().numpy()

# Visualize a subset of feature maps (you can choose which channels to display)
num_feature_maps = features.shape[0]  # This should be 512
channels_to_display = min(10, num_feature_maps)  # Display up to 10 feature maps

# Plot the first 10 feature maps
fig, axes = plt.subplots(2, 5, figsize=(15, 6))  # 2 rows and 5 columns
for i in range(channels_to_display):
    ax = axes[i // 5, i % 5]
    ax.imshow(features[i], cmap='viridis')  # Using 'viridis' for better visibility
    ax.axis('off')
    ax.set_title(f"Feature Map {i+1}")

plt.tight_layout()
plt.show()

# Print the shape of the feature tensor
print("Feature Shape:", features.shape)
```
