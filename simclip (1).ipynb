```python
import os
from PIL import Image
import pandas as pd
from transformers import CLIPProcessor, CLIPModel
import torch
```

```python
from tqdm import tqdm
```

```python
import zipfile as zf
files = zf.ZipFile("/content/drive/My Drive/facebookds.zip", 'r')
files.extractall('memes')
files.close()
```

```python
# Step 1: Load SimCLIP Model
model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

```python
# Step 2: Define Function to Get SimCLIP Embeddings
def get_clip_embeddings(image, text):
    inputs = processor(text=[text], images=image, return_tensors="pt", padding=True)
    text_emb = model.get_text_features(**inputs)
    image_emb = model.get_image_features(**inputs)
    return text_emb, image_emb
```

```python
# Step 3: Load JSONL Data
# Path to the JSONL file and memes folder
jsonl_file_path = "/content/memes/data/ds.jsonl"  # Replace with your file path
memes_folder = "/content/memes/data/"  # Folder containing meme images
```

```python
data = []
with open(jsonl_file_path, "r") as file:
    for line in file:
        data.append(json.loads(line))
```

```python
# Step 4: Generate Embeddings
def preprocess_data(entry):
    image_path = os.path.join(memes_folder, entry['img'])
    image = Image.open(image_path).convert("RGB")
    text = entry['text']
    text_emb, image_emb = get_clip_embeddings(image, text)
    return torch.cat((text_emb, image_emb), dim=1).squeeze(0)
```

```python
# Process JSONL data directly
embeddings = []
labels = []
for entry in tqdm(data, desc="Processing memes"):
    try:
        embedding = preprocess_data(entry)
        embeddings.append(embedding)
        labels.append(entry['label'])
    except Exception as e:
        print(f"Error processing {entry['img']}: {e}")
```

```python
def get_clip_embeddings(image, text):
    # Preprocess image and text separately
    image_inputs = processor(images=image, return_tensors="pt", padding=True)
    text_inputs = processor(text=[text], return_tensors="pt", padding=True)

    # Extract features separately
    text_emb = model.get_text_features(input_ids=text_inputs.input_ids, attention_mask=text_inputs.attention_mask)
    image_emb = model.get_image_features(pixel_values=image_inputs.pixel_values)

    return text_emb, image_emb
```

```python









# Step 4: Generate Embeddings
def preprocess_data(row):
    image_path = os.path.join(memes_folder, row['image_name'])
    image = Image.open(image_path).convert("RGB")
    text = row['extracted_text']
    text_emb, image_emb = get_clip_embeddings(image, text)
    return torch.cat((text_emb, image_emb), dim=1).squeeze(0)

# Apply the embedding generation
data_df['combined_embedding'] = data_df.apply(preprocess_data, axis=1)

# Step 5: Save Processed Data
# Save combined embeddings and labels for training
embeddings = torch.stack(data_df['combined_embedding'].tolist())
labels = torch.tensor(data_df['label'].tolist())

torch.save(embeddings, "/content/embeddings.pt")
torch.save(labels, "/content/labels.pt")

print("Embeddings and labels saved successfully!")
```
