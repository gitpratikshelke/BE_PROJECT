```python

class MemeDataset(Dataset):
    def __init__(self, json_path, meme_dir):
        with open(json_path, 'r') as f:
            self.data = json.load(f)

        self.meme_dir = meme_dir
        self.transform_clip = clip_processor.image_processor
        self.transform_siglip = siglip_processor.image_processor

        # Preprocess dataset
        self.data = [
            {
                'img_path': os.path.join(self.meme_dir, item['img']),
                'text': f"{item['text']} {item.get('caption', '')}".strip(),
                'label': item['label']
            }
            for item in self.data if os.path.exists(os.path.join(self.meme_dir, item['img']))
        ]

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image = Image.open(item['img_path']).convert('RGB')

        # Apply transformations
        clip_image = self.transform_clip(image)
        siglip_image = self.transform_siglip(image)

        return {
            'clip_image': clip_image['pixel_values'][0],
            'siglip_image': siglip_image['pixel_values'][0],
            'text': item['text'],
            'label': torch.tensor(item['label'], dtype=torch.long)
        }
```

```python
import os
import json
import torch
import numpy as np
import pandas as pd
from PIL import Image
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from transformers import CLIPModel, CLIPProcessor, CLIPTokenizer
from transformers import SiglipModel, SiglipProcessor
```

```python
class MemeDataset(Dataset):
    def __init__(self, json_path, meme_dir, clip_processor, siglip_processor):
        with open(json_path, 'r') as f:
            self.data = json.load(f)
        self.meme_dir = meme_dir
        self.clip_processor = clip_processor
        self.siglip_processor = siglip_processor
        self._validate_and_process_data()

    def _validate_and_process_data(self):
        valid_data = []
        for item in self.data:
            img_path = os.path.join(self.meme_dir, item['img'])
            if not os.path.exists(img_path):
                continue
            text = f"{item.get('text', '')} {item.get('caption', '')}".strip()
            if 'label' not in item:
                continue
            valid_data.append({
                'img_path': img_path,
                'text': text,
                'label': item['label']
            })
        self.data = valid_data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        image = Image.open(item['img_path']).convert('RGB')

        # Process images
        clip_img = self.clip_processor(
            images=image, return_tensors="pt")['pixel_values'].squeeze(0)
        siglip_img = self.siglip_processor(
            images=image, return_tensors="pt")['pixel_values'].squeeze(0)

        return {
            'clip_image': clip_img,
            'siglip_image': siglip_img,
            'text': item['text'],
            'label': torch.tensor(item['label'], dtype=torch.long)
        }
```

```python
class MultimodalEnsemble(nn.Module):
    def __init__(self, clip_model, siglip_model, num_classes=2):
        super().__init__()
        self.clip = clip_model
        self.siglip = siglip_model

        # Freeze models
        for param in self.clip.parameters():
            param.requires_grad = False
        for param in self.siglip.parameters():
            param.requires_grad = False

        self.clip_img_fc = nn.Sequential(
            nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.2))
        self.siglip_img_fc = nn.Sequential(
            nn.Linear(768, 256), nn.ReLU(), nn.Dropout(0.2))
        self.text_fc = nn.Sequential(
            nn.Linear(512, 256), nn.ReLU(), nn.Dropout(0.2))

        self.classifier = nn.Linear(256*3, num_classes)

    def forward(self, clip_img, siglip_img, text_ids):
        # Image features
        clip_img_feat = self.clip.vision_model(clip_img).pooler_output
        siglip_img_feat = self.siglip.vision_model(siglip_img).pooler_output

        # Text features
        text_feat = self.clip.text_model(text_ids).pooler_output

        # Project features
        clip_img_embed = self.clip_img_fc(clip_img_feat)
        siglip_img_embed = self.siglip_img_fc(siglip_img_feat)
        text_embed = self.text_fc(text_feat)

        # Concatenate and classify
        combined = torch.cat([clip_img_embed, siglip_img_embed, text_embed], dim=1)
        return self.classifier(combined)
```

```python

def train_epoch(model, loader, criterion, optimizer, device, clip_tokenizer):
    model.train()
    total_loss = 0.0
    for batch in loader:
        optimizer.zero_grad()

        # Process inputs
        clip_imgs = batch['clip_image'].to(device)
        siglip_imgs = batch['siglip_image'].to(device)
        texts = batch['text']
        labels = batch['label'].to(device)

        # Tokenize text
        text_inputs = clip_tokenizer(
            texts, padding=True, truncation=True,
            return_tensors="pt", max_length=128
        ).to(device)

        # Forward pass
        outputs = model(clip_imgs, siglip_imgs, text_inputs.input_ids)
        loss = criterion(outputs, labels)

        # Backward pass
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    return total_loss / len(loader)
```

```python
def evaluate_model(model, loader, device, clip_tokenizer):
    model.eval()
    all_preds = []
    all_labels = []
    with torch.no_grad():
        for batch in loader:
            clip_imgs = batch['clip_image'].to(device)
            siglip_imgs = batch['siglip_image'].to(device)
            texts = batch['text']
            labels = batch['label'].numpy()

            text_inputs = clip_tokenizer(
                texts, padding=True, truncation=True,
                return_tensors="pt", max_length=128
            ).to(device)

            outputs = model(clip_imgs, siglip_imgs, text_inputs.input_ids)
            preds = torch.argmax(outputs, dim=1).cpu().numpy()

            all_preds.extend(preds)
            all_labels.extend(labels)

    print(classification_report(all_labels, all_preds))
    print("Confusion Matrix:")
    print(confusion_matrix(all_labels, all_preds))
```

```python
def main():
    # Configuration
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    BATCH_SIZE = 16
    LR = 1e-4
    EPOCHS = 10

    # # Initialize models and processors
    # clip_model = CLIPModel.from_pretrained("openai/clip-vit-base-patch32")
    # clip_processor = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
    # clip_tokenizer = CLIPTokenizer.from_pretrained("openai/clip-vit-base-patch32")

    # siglip_model = SiglipModel.from_pretrained("google/siglip-base-patch16-224")
    # siglip_processor = SiglipProcessor.from_pretrained("google/siglip-base-patch16-224")

    # Dataset and Loaders
    dataset = MemeDataset(
        json_path='/content/balanced_new.json',
        meme_dir='/content/meme_dataset/data',
        clip_processor=clip_processor,
        siglip_processor=siglip_processor
    )

    train_data, val_data = train_test_split(dataset, test_size=0.2, random_state=42)
    train_loader = DataLoader(train_data, batch_size=BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_data, batch_size=BATCH_SIZE)

    # Model setup
    model = MultimodalEnsemble(clip_model, siglip_model).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=LR)

    # Training loop
    for epoch in range(EPOCHS):
        train_loss = train_epoch(model, train_loader, criterion, optimizer, device, clip_tokenizer)
        print(f"Epoch {epoch+1}/{EPOCHS} | Train Loss: {train_loss:.4f}")
        print("Validation Metrics:")
        evaluate_model(model, val_loader, device, clip_tokenizer)

    # Save model
    torch.save(model.state_dict(), 'multimodal_toxicity_model.pth')

if __name__ == "__main__":
    main()
```

```python
import torch.nn as nn
from transformers import AutoModel

class SigLIPClassifier(nn.Module):
    def __init__(self, model_name):
        super(SigLIPClassifier, self).__init__()
        self.siglip = AutoModel.from_pretrained(model_name)

        # Access vision_config through the main model's config
        hidden_dim = self.siglip.config.vision_config.hidden_size  # Corrected line

        self.fc = nn.Linear(hidden_dim, 1)  # Binary classification

    def forward(self, inputs):
        outputs = self.siglip(**inputs).last_hidden_state[:, 0, :]  # CLS token
        logits = self.fc(outputs)
        return torch.sigmoid(logits)
```
